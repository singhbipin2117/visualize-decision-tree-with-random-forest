```python
# load required library
import numpy as np
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import  plot_tree
```

```python
def save_decision_trees_as_png(clf, iteration, feature_name, target_name):
    plt.figure(figsize=[10.0,10.0])
    plot_tree(clf, filled=True, feature_names=feature_name, class_names=target_name, node_ids=True)
    plt.show()
    print("Decision Tree {}".format(iteration + 1))
```

```python
model = RandomForestClassifier(n_estimators=5)
model.fit(iris.data, iris.target)
```

```python
model.estimators_
```

```python
model.estimators_[0]
```

```python
len(model.estimators_)
```

```python
for i in range(len(model.estimators_)):
    save_decision_trees_as_png(model.estimators_[i], i, features_name, target_name)
```
